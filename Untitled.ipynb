```python
import psstdata
import pandas as pd

data = psstdata.load()
# for utterance in data.train[:4]:
    
#     # The key ingredients
#     utterance_id = utterance.utterance_id
#     transcript = utterance.transcript
#     correctness = "Y" if utterance.correctness else "N"
#     filename_absolute = utterance.filename_absolute

#     print(f"{utterance_id:26s} {transcript:26s} {correctness:11s} {filename_absolute}")
```

```python
d_train=data.train
df_train = pd.DataFrame()
v_utterance = [] ; v_session = [] ; v_test = [] ; v_prompt = [] ; v_transcription = [] 
v_correctness = []  ; v_aq_index = [] ; v_duration = [] ; v_filename = []
for i in range(len(d_train)):
    v_utterance.append(d_train.utterances[i].utterance_id)
    v_session.append(d_train.utterances[i].session)
    v_test.append(d_train.utterances[i].test)
    v_prompt.append(d_train.utterances[i].prompt)
    v_transcription.append(d_train.utterances[i].transcript)
    v_correctness.append(d_train.utterances[i].correctness)
    v_aq_index.append(d_train.utterances[i].aq_index)
    v_duration.append(d_train.utterances[i].duration_frames)
    v_filename.append(d_train.utterances[i].filename)

df_train=df_train.assign(utterance_id=v_utterance)
df_train=df_train.assign(session=v_session)
df_train=df_train.assign(test=v_test)
df_train=df_train.assign(prompt=v_prompt)
df_train=df_train.assign(transcription=v_transcription)
df_train=df_train.assign(correctness=v_correctness)
df_train=df_train.assign(aq_index=v_aq_index)
df_train=df_train.assign(duration=v_duration)
df_train=df_train.assign(filename=v_filename)

df_train['phonemes'] = (df_train['transcription'].str.split(" "))
```
